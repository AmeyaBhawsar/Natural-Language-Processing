```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

```

```python
img1 = cv2.imread('images/eiphel1.png', 0)
img2 = cv2.imread('images/eiphel2.png', 0)

figure, ax = plt.subplots(1,2, figsize=(16,8))

ax[0].imshow(img1, cmap='gray')
ax[1].imshow(img2, cmap='gray');
```

```python
# Sift
sift = cv2.SIFT_create()
```

```python
keypoints_1, descripters_1 = sift.detectAndCompute(img1, None)
keypoints_2, descripters_2 = sift.detectAndCompute(img2, None)

len(keypoints_1), len(keypoints_2)
```

```python
# Features Matching

bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck = True)

matches = bf.match(descripters_1, descripters_2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:20], img2, flags = 2)

plt.imshow(img3);
```
