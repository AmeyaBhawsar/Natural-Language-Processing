```python
import nltk
from nltk.tokenize import word_tokenize
import string

import requests
from bs4 import BeautifulSoup
import re
```

### Question 1

```python
#Question 1
sent= "They told that their ages are 23, 25 and 29 respectively."
```

```python
#Tokenizing the sentence
tokenized_words = word_tokenize(sent)
tokenized_words
```

```python
#1. Find the average of ages.
num=[int(word) for word in tokenized_words if word.isnumeric()]
average_age = sum(num) / len(num)
print("the average of ages:",average_age)
```

```python
#2. Count total number of words starts with 't'/'T'

t_words_count = [word for word in tokenized_words if word.startswith('t') or word.startswith('T')]
print(t_words_count)
print("\n Count of words starts with 't'/'T': ",len(t_words_count))
```

```python
# 3. Count total punctuation symbols.

punctuation_count = [char for char in sent if char in string.punctuation]
print(punctuation_count)
print("\n Total punctuation symbols: ",len(punctuation_count))
```

```python
#4. Print the words in capital letter which ends with 'd'

words_ending_d = [word.upper() for word in tokenized_words if word.lower().endswith('d')]
print("\n Words in capital letter which ends with 'd': \n ",words_ending_d)
```

```python
#5. Merge all the words and print (no punctuations)

words_no_punctuation = [word for word in tokenized_words if word not in string.punctuation]
merged_words = ' '.join(words_no_punctuation)
print(words_no_punctuation)
print("\n Sentence with no punctuations: \n ",merged_words)
```

### Question 2

```python
#Question 2

# Fetch the Wikipedia page of the Indian flag
url = "https://en.wikipedia.org/wiki/Flag_of_India"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract the text from the page
text = soup.get_text()
text
```

```python
# Tokenize the words
tokenized_words = nltk.word_tokenize(text)
print(tokenized_words)
```

```python
#1 Count the total number of words
w_count = len(tokenized_words)
print("Total number of words is:",w_count)
```

```python
# 2. Count percentage of punctuation marks

# Count of no of punctuation symbols
punctuations = [char for char in text if char in string.punctuation]

#calculate percentage
punctuation_percentage = (len(punctuations) / len(tokenized_words)) * 100

print(punctuations)
print("\nCount of punctuation marks",len(punctuations))
print("\n Percentage of punctuation marks is: ",punctuation_percentage)
```

```python
#3.  Print the total number of lines
lines_count = text.count('\n')
print("Total number of lines are: ",lines_count)
```

```python
# 4. how many words are in title case
title_case_count = [word for word in tokenized_words if word.istitle()]
print(title_case_count)
print("\n Total number of words in titile case: ",len(title_case_count))
```

```python
# 5. Remove the numbers and print the text
words_without_numbers = [word for word in tokenized_words if not re.search(r'\d', word)]
    
# Join the words back into a single string
text_without_numbers = ' '.join(words_without_numbers)

print("Word without numbers:\n",text_without_numbers)
```
