```python
!pip install wikipedia==1.4.0
#Question 1
import wikipedia
nasa_data = wikipedia.page("NASA").content
```

```python

!pip install nltk==3.8.1

import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

```

```python
# 1. Apply POS tagging on this data
tokens = nltk.word_tokenize(nasa_data)
tagged_tokens = nltk.pos_tag(tokens)
tagged_tokens
```

```python
# 2. Remove punctuation symbols
import string

no_punctuation = [word for word in tokens if word not in string.punctuation ]
no_punctuation
```

```python
# 3. Delete Stopwords
nltk.download('stopwords')
stop_words = set(nltk.corpus.stopwords.words('english'))

no_stopwords = [word for word in tokens if  word.lower() not in stop_words]
no_stopwords
```

```python
# 4. Perform morphological analysis on all nouns
nouns=[]
for i in tagged_tokens:
  if i[1].startswith('N'):
        nouns.append(i[0])
print(nouns)
```

```python
# performed morphologyical analysis
for w in nouns:
  w = Word(w, language="en")
  print(w, '\t\t', w.morphemes)
```

```python
# 5. Find top three words in the data
# for calcualting the frquency distribution
from nltk.probability import FreqDist
```

```python
# Get the top three words
freq_dist = FreqDist(tokens)

top_three_words = freq_dist.most_common(3)
top_three_words
```
