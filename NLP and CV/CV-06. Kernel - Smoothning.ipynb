```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Averaging Kernel

```python
image=cv2.imread("images/lenna.png")

#Define a 5X5 averaging kernel
kernel=np.ones((5,5),np.float32)/25

#Apply the kernel to smooth the image
smoothed_image=cv2.filter2D(image,-1,kernel)

#show the orginal and smoothed images
cv2.imshow("Original",image)
cv2.imshow("Smoothed",smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_image=np.hstack([image,smoothed_image]) # combines both images in single frame using hstack
                                            #i.e horizontal stack
cv2.imshow("Both images",new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Gaussian kernel


```python
kernel_size=7 #adjust as needed
sigma=0.35*((kernel_size-1)*0.5-1)+0.9 # gaussian standard deviation
kernel1=cv2.getGaussianKernel(kernel_size,sigma)
 
```

```python
sigma
```

```python
kernel1
```

```python
#Apply the kernel to smooth the image
smoothed_image=cv2.filter2D(image,-1,kernel1)

#show the orginal and smoothed images
cv2.imshow("Original",image)
cv2.imshow("Smoothed",smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Median Kernel

```python
gray = cv2.imread('images/lenna.png', 0)
```

```python
gray.shape
```

```python
import matplotlib.pyplot as plt
```

```python
kernel_size = 5
median = cv2.medianBlur(gray, kernel_size)
plt.figure(figsize=(10,6))
plt.subplot(121), plt.imshow(gray, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title('Median Filtered Image')
plt.show()
```

```python
gray = cv2.imread('images/lenna.png')
kernel_size = 15 # more the size of kernel more blurness 
median = cv2.medianBlur(gray, kernel_size)
plt.figure(figsize=(10,6))
plt.subplot(121), plt.imshow(gray, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title('Median Filtered Image')
plt.show()
```

# Thresholding

```python
image=cv2.imread("images/lenna.png",0)

binary_image=image>100
```

```python
binary_image
```

```python
#show the orginal and binary images
plt.subplot(1,2,1)
plt.title('original image')
plt.imshow(image, cmap='gray')
plt.subplot(1,2,2)
plt.title('binary image')
plt.imshow(binary_image, cmap='gray')
```

```python
binary_image.dtype
```

```python
binary_image=binary_image.astype(np.uint8)*255
```

```python
cv2.imshow("Binary Image",binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
binary_image
```

## Inverese binary threshold

```python
inv_binary_image=image<200
```

```python
#show the orginal and binary images
plt.subplot(1,2,1)
plt.title('original image')
plt.imshow(image, cmap='gray')
plt.subplot(1,2,2)
plt.title('Inverse binary image')
plt.imshow(inv_binary_image, cmap='gray')
```

```python
ret, thresh1 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY_INV)
#show the orginal and binary images
plt.figure(figsize=(16,8))
plt.subplot(1,3,1)
plt.title('original image')
plt.imshow(image, cmap='gray')
plt.subplot(1,3,2)
plt.title('Binary image')
plt.imshow(thresh1, cmap='gray')
plt.subplot(1,3,3)
plt.title('Inverse Binary image')
plt.imshow(thresh2, cmap='gray')
```
