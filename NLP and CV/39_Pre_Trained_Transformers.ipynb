```python
from transformers import pipeline
```

```python
#TEXT CLASSIFIER
classifier = pipeline("text-classification")
```

```python
import pandas as pd
text='It was possible for him easily. So he did it.'
result=classifier(text)
df=pd.DataFrame(result)
df
```

```python
text='I did not like it'
result=classifier(text)
df1=pd.DataFrame(result)
df1
```

Question Answering


```python
qa=pipeline("question-answering")
```

```python
text='After K alphabet M comes'
question='what comes after K?'
result=qa(question=question,context=text)
result
```

Named Entity Recognition

```python
ner=pipeline('token-classification')
```

```python
text=''' Hugging Face, Inc. is a French-American company incorporated under the Delaware
 General Corporation Law and based in New York City that develops computation tools for building applications using machine learning'''
result=ner(text)
pd.DataFrame(result)
```

Conversation

```python
from transformers import Conversation
converse=pipeline('conversational')
```

```python
Conversation=Conversation(" It was my first day to job. Any suggesations?")
response=converse([Conversation])
```

```python
response[1]['content']
```

Visual Question Answering

```python
from PIL import Image
vqa_pipeline=pipeline("visual-question-answering")
```

```python
image = Image.open('/content/images.jfif')
image
```

```python
question='Who is there in image?'
vqa_pipeline(image,question,top_k=3)
```

```python
question='What is color in image?'
vqa_pipeline(image,question,top_k=3)[0]
```

```python
question='What is car name?'
vqa_pipeline(image,question,top_k=3)
```
