```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
int_to_char= dict(enumerate(alphabets)) 
```

```python
int_to_char
```

```python
char_to_int=dict((v,i) for i,v in enumerate(alphabets))
```

```python
char_to_int
```

#### Data Prepration

```python
seq_length=1
x=[]
y=[]

for i in range(len(alphabets)-seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
    print(seq_in,'->', seq_out)
```

```python
x
```

```python
y
```

### Prepare the output variable

```python
from keras.utils import to_categorical
```

```python
y_new  = to_categorical(y)  #One Hot Encoding
```

```python
y_new
```

#### Prepare the input variables

```python
x=np.reshape(x,(25,1,1))
```

```python
x
```

```python
x.shape
```

```python
x_new = x / 25  #Normalizing data
```

```python
x_new
```

```python
x_new.shape
```

```python
y_new.shape
```

### Create the model 

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(1,1), activation='relu'))
model.add(LSTM(32))
model.add(Dense(26, activation='softmax'))
```

### Compile the model

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Train the model

```python
model.fit(x_new, y_new, epochs=1000, batch_size=1)
```

```python
model.evaluate(x_new,y_new,batch_size=1)
```

```python
new = 'O'
new = char_to_int[new]
new = np.reshape(new,(1,1,1))
new = new/25
pred = model.predict(new)
pred = np.argmax(pred)
int_to_char[pred]
```

```python
seq_length=3
x=[]
y=[]

for i in range(len(alphabets)-seq_length):
    seq_in = alphabets[i:i + seq_length]
    seq_out = alphabets[i+seq_length]
    x += [[char_to_int[char] for char in seq_in]]
    y += [char_to_int[seq_out]]
    print(seq_in,'->', seq_out)
```

```python
x
```

```python
x=np.reshape(x,(len(x),seq_length,1))
print(x.shape)

x_new = x / 25  #Normalizing data
y_new=to_categorical(y)
print(y_new.shape)
```

```python
model1 = Sequential()

model1.add(Dense(25, input_shape=(3,1), activation='relu'))
model1.add(LSTM(128))
model1.add(Dense(26, activation='softmax'))
```

```python
model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model1.fit(x_new, y_new, epochs=1000, batch_size=1)
```

```python
model1.evaluate(x_new,y_new,batch_size=1)
```

```python
new = 'GOA'
new = [char_to_int[char] for char in new]
new = np.reshape(new,(1,3,1))
new = new/25
pred = model1.predict(new)
pred = np.argmax(pred)
int_to_char[pred]
```
