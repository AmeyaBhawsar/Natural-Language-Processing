```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('C:/Users/Administrator.DAI-PC2/NLP and CV/images/cat_damaged.png')

height, width = image.shape[:-1]
height, width
```

```python
cv2.imshow('Damaged Cat', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask = np.zeros([height, width, 3], dtype=np.uint8)
```

```python
for i in range (height):
    for j in range(width):
        if image[i,j].sum() == 0:
            mask[i,j] = [255,255,255]
```

```python
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(mask, cmap='gray')
```

```python
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
```

```python
# Impaint
dst = cv2.inpaint(image, mask,3,cv2.INPAINT_NS)
```

```python
cv2.imshow('Reconstructed Cat', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
