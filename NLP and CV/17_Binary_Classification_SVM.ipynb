### Prepration of data

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
sent='Hello friends! How are you ? I like Python Programming '
```

```python
#1. tokenize the text
tokens1=word_tokenize(sent)
```

```python
tokens1
```

```python
#2. remove punctuations
tokens2=[token for token in tokens1 if token.isalnum()]
```

```python
tokens2
```

```python
# 3. Remove the stopwords
swords=stopwords.words('english')
```

```python
token3=[token for token in tokens2 if token.lower() not in swords]
token3
```

```python
#4. Remove the suffixes
pos=PorterStemmer()
token4=[pos.stem(word) for word in token3]
print(token4)
```

```python
def clean_text(sent):
    tokens1=word_tokenize(sent)
    tokens2=[token for token in tokens1 if token.isalnum()]
    token3=[token for token in tokens2 if token.lower() not in swords]
    token4=[pos.stem(word) for word in token3]
    return token4

```

```python
clean_text(sent)
```

### TF*IDF Vectorizer

```python
#import the class
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#create the object
tfidf=TfidfVectorizer(analyzer=clean_text)
```

```python
#apply
x_new=tfidf.fit_transform(x)
```

```python
x_new.shape
```

### Cross- Validation

```python
from sklearn.model_selection import train_test_split
```

```python
#seperate the training and testing data
x_train,x_test,y_train,y_test=train_test_split(x_new,y,random_state=0)  #defaultsize 75%,25% .ie testsize=0.25
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

#

#### Build the model using SVM

```python
# import the class
from sklearn.svm import SVC
```

```python
#create the object
svm=SVC(kernel='linear',random_state=0)
```

```python
%%time
svm.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
y_pred=svm.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
```
